## Random Walk

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().strftime('%s'))
def getLinks(articleUrl):
    html = urlopen(f'http://en.wikipedia.org{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('/wiki/Qarnayel')
while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

# Set a random seed based on the current time to make sure the random selection is different every time.
random.seed(datetime.datetime.now().strftime('%s'))

# Function to get all valid links from a Wikipedia article
def getLinks(articleUrl):
    html = urlopen(f'http://en.wikipedia.org{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    # Find all links that match the Wikipedia article pattern (ignores links to external resources)
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

# Start from a specific Wikipedia article (Qarnayel in this case)
links = getLinks('/wiki/Qarnayel')
visited = set()  # Keep track of visited articles to avoid revisiting

# Limit the crawl to 50 links
counter = 0
max_links = 50

while len(links) > 0 and counter < max_links:
    # Randomly choose a link to follow
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']

    # Check if we haven't visited this article before
    if newArticle not in visited:
        print(f"Article {counter + 1}: {newArticle}")
        visited.add(newArticle)
        links = getLinks(newArticle)  # Get the links from the new article
        counter += 1  # Increment the counter
    else:
        # If the link was already visited, just pick a new one
        links = getLinks(newArticle)  # Get the links from the new article again
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
def getLinks(pageUrl):

    html = urlopen(f'http://en.wikipedia.org{pageUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #We have encountered a new page
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
getLinks('')
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError  # Import HTTPError to catch 404 errors
from bs4 import BeautifulSoup
import re

pages = set()
page_counter = 0  # Counter to track the number of pages we've visited
max_pages = 20  # Limit to 20 pages

def getLinks(pageUrl):
    global page_counter  # Access the global counter variable
    if page_counter >= max_pages:  # Stop if we've reached the max page limit
        return

    try:
        html = urlopen(f'http://en.wikipedia.org{pageUrl}')
        bs = BeautifulSoup(html, 'html.parser')

        # Find all links that match Wikipedia internal links (links starting with /wiki/)
        for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
            if 'href' in link.attrs:
                newPage = link.attrs['href']
                if newPage not in pages:
                    # We have encountered a new page
                    print(newPage)  # Print the new page's link
                    pages.add(newPage)  # Add the link to the visited pages set
                    page_counter += 1  # Increment the page counter
                    getLinks(newPage)  # Recursively get links from the new page
                    if page_counter >= max_pages:  # Check again if we've reached the limit
                        return

    except HTTPError as e:
        # Catch HTTPError (like 404 or other errors) and print a message, but continue the script
        print(f"HTTPError encountered on {pageUrl}: {e}")

# Start the process from the Wikipedia main page (or any other page URL)
getLinks('')

```

## Crawling across the Internet

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random

#Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

#Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    bs = BeautifulSoup(urlopen(startingPage), 'html.parser')
    externalLinks = getExternalLinks(bs, startingPage)
    if not len(externalLinks):
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        return getRandomExternalLink(random.choice(internalLinks))
    else:
        return random.choice(externalLinks)

def followExternalOnly(startingSite):
    externalLink = getRandomExternalLink(startingSite)
    print(f'Random external link is: {externalLink}')
    followExternalOnly(externalLink)


followExternalOnly('https://www.oreilly.com/')
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError  # To handle HTTP errors
from bs4 import BeautifulSoup
import re

allExtLinks = []  # List to collect external links
allIntLinks = []  # List to collect internal links
max_links = 30    # Maximum number of links to collect

internal_counter = 0  # Counter for internal links
external_counter = 0  # Counter for external links

def getInternalLinks(bs, url):
    internalLinks = []
    # Find all internal links (those starting with '/')
    for link in bs.find_all('a', href=re.compile('^(/)')):
        if 'href' in link.attrs:
            internalLinks.append(url + link.attrs['href'])
    return internalLinks

def getExternalLinks(bs, url):
    externalLinks = []
    # Find all external links (those that start with 'http')
    for link in bs.find_all('a', href=re.compile('^(http|https)')):
        if 'href' in link.attrs:
            externalLinks.append(link.attrs['href'])
    return externalLinks

def getAllExternalLinks(url):
    global internal_counter, external_counter

    if internal_counter + external_counter >= max_links:
        return  # Stop if we have collected enough links

    try:
        bs = BeautifulSoup(urlopen(url), 'html.parser')  # Open the URL and parse HTML

        # Get internal and external links from the page
        internalLinks = getInternalLinks(bs, url)
        externalLinks = getExternalLinks(bs, url)

        # Process external links
        for link in externalLinks:
            if link not in allExtLinks and external_counter < max_links:
                allExtLinks.append(link)
                external_counter += 1
                print(f"External Link {external_counter}: {link}")

        # Process internal links
        for link in internalLinks:
            if link not in allIntLinks and internal_counter < max_links:
                allIntLinks.append(link)
                internal_counter += 1
                print(f"Internal Link {internal_counter}: {link}")
                if internal_counter + external_counter < max_links:
                    getAllExternalLinks(link)  # Recurse on internal links

    except HTTPError as e:
        print(f"HTTPError encountered on {url}: {e}")  # Handle errors and continue

# Start scraping from the given URL (make sure it’s valid)
start_url = 'https://www.oreilly.com/'
getAllExternalLinks(start_url)
```
